```python
# Importing the required packages here

import numpy as np
import pandas as pd
import seaborn as sns
import chardet
import missingno

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud
```

```python
# Identify the encoding used
with open('datasets/UsArrests.csv', 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result
```

```python
# Loading the dataset using the identified encoding

arrests_df = pd.read_csv('datasets/UsArrests.csv', encoding = "utf-8-sig")

arrests_df.head()
```

```python
# Missing matrix, to advise of any values missing in the data
missingno.matrix(arrests_df, figsize = (20, 10))
print("From this graph, we can understand that there are no entries missing data")
```

```python
# Print statement to advise of data types
print(arrests_df.dtypes)
```

```python
# 50 States in a word cloud
states = arrests_df["City"]

text = " ".join(states)

wordcloud = WordCloud(width = 400, height = 400, margin = 5).generate(text)
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.margins(x = 0, y = 0)
plt.show
```

```python
# Data sorted by Murder values
print("Sorted values from Murder column")
arrests_df.sort_values("Murder")
```

```python
# Data sorted by Assault values
print("Sorted values from Assault column")
arrests_df.sort_values("Assault")
```

```python
# Data sorted by Rape values
print("Sorted values from Rape column")
arrests_df.sort_values("Rape")
```

```python
# Data sorted by UrbanPop values
print("States with the highest UrbanPop")
arrests_df.sort_values("UrbanPop").tail()
```

```python
arrests_df.sort_values("UrbanPop").head()
```

```python
murder = arrests_df["Murder"]
city = arrests_df["City"]

plt.bar(city, murder)
plt.title("Number of Murders per state")
plt.xticks(rotation = 90)
plt.show
```

```python
assault = arrests_df["Assault"]
city = arrests_df["City"]

plt.bar(city, assault)
plt.title("Number of Assaults per state")
plt.xticks(rotation = 90)
plt.show
```

```python
rape = arrests_df["Rape"]
city = arrests_df["City"]

plt.bar(city, rape)
plt.title("Number of Rapes per state")
plt.xticks(rotation = 90)
plt.show
```

```python
pop = arrests_df["UrbanPop"]

plt.plot(city, pop)
plt.title("Urban Pop")
plt.xticks(rotation = 90)
plt.show()
```
